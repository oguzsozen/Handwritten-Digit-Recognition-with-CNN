# MNIST Dataset

# Dataset'in İndirilmesi, İncelenmesi ve Ön işlem

```python
# Dataset'in çekilmesi, Numpy array'larla işlem yapabilmek ve target'ları
# katogerize etmek için gerekli kütüphanelerin import edilmesi
import numpy as np
from keras.datasets import mnist
from tensorflow.keras.utils import to_categorical
```

```python
# mnist dataset'in indirilmesi
(train_images, train_labels), (test_images, test_labels) = mnist.load_data()
```

```python
train_labels.shape
```

```python
test_images.shape
```

```python
test_labels.shape
```

```python
# train ve test verilerinin yeniden boyutlandırılması ve normalize edilmesi
train_images = train_images.reshape((60000, 28, 28, 1))
train_images = train_images.astype('float32') /255
```

```python
test_images = test_images.reshape((10000, 28, 28, 1))
test_images = test_images.astype('float32') /255
```

```python
# Label'ların to_categorical fonksiyonu sayesinde etiketlendirilmesi
train_labels = to_categorical(train_labels)
test_labels = to_categorical(test_labels)
```

# CNN'li Modeli Oluşturma, Eğitme ve Başarımı Değerlendirme

```python
# Model için gerekli kütüphanelerin import edilmesi
from keras import layers
from keras import models
```

```python
# Kurulan basit CNN modeli
model = models.Sequential()

model.add(layers.Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1)))
model.add(layers.MaxPool2D((2, 2)))
model.add(layers.Conv2D(64, (3, 3), activation='relu'))
model.add(layers.MaxPool2D((2, 2)))
model.add(layers.Conv2D(64, (3, 3), activation='relu'))

model.add(layers.Flatten())

model.add(layers.Dense(64, activation='relu'))
model.add(layers.Dense(10, activation='softmax'))
```

```python
model.summary()
```

```python
# Modelin compile edilmesi
model.compile(
    optimizer='rmsprop',
    loss='categorical_crossentropy',
    metrics=['accuracy'])
```

```python
# Kurulan modelin eğitilmesi
model.fit(train_images,
          train_labels,
          epochs=5,
          batch_size=64)
```

```python
# Son olarak test verisi ile başarımının ölçülmesi
test_loss, test_acc = model.evaluate(test_images, test_labels)
```
